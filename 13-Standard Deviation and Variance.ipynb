```python
from typing import TypeVar


T = TypeVar('T')

def pr(val : T , title:str|None = None) : # type: ignore
    if title != None :
        print(f" ---------------------- {title} ---------------------- ")
    print('type: ' , type(val))
    print(val , '\n')
    # return val
```

```python
import numpy as np # type: ignore
import numpy.typing as npt # type: ignore
```

```python
# this lesson is about standard deviation and variance
# standard deviation is the square root of variance
high_var_array  = np.array([1,100,200,300,4000,5000])
low_var_array = np.array([2,4,6,8,10])
pr(high_var_array.var() , 'high_var_array.var()' )
pr(np.var(high_var_array) , 'np.var(high_var_array)' )
pr(low_var_array.var() , 'low_var_array.var()' )
pr(np.var(low_var_array) , 'np.var(low_var_array)' )
pr(high_var_array.std() , 'high_var_array.std()' )
pr(np.std(high_var_array) , 'np.std(high_var_array)' )
pr(low_var_array.std() , 'low_var_array.std()' )
pr(np.std(low_var_array) , 'np.std(low_var_array)' )
```

```python
# remember the standard deviation is the average distance that the values move away from it's mean.
pr(np.mean(high_var_array) , 'np.mean(high_var_array)' )
pr(np.std(high_var_array) , 'np.std(high_var_array)' )
pr(np.mean(low_var_array) , 'np.mean(low_var_array)' )
pr(np.std(low_var_array) , 'np.std(low_var_array)' )
```

```python
# let's try to import the data
%matplotlib inline
import matplotlib.pyplot as plt
plt.hist(high_var_array)
plt.title('high variance array')
plt.show()
```
